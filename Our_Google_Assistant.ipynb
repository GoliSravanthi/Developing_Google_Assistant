To do this task successfully,we need to follow the following steps ..
1.   taking question as an input
2.   classifying question whether it belongs too device operations or not.
3.  check whether given query can be searched in internet or not.
4.  if it related to search internet, get it from google gemini model
5.   convert google text to speech ie,gTTS









**STEP 1:: taking question as an input.**

```python
def ask_question(name):
  query="hey "+name +" How can I help You ?"
  query=input(query)
  return query
```

```python
ask_question("sathya")
```

**Step 2:: Classifying the question**

```python
def classify_query(query):
  device_lst=['set an alarm',' remainder','message ','call']   #making a list of device operations
  device=False
  for i in device_lst:
    if i in query:
      device=True
  if device:
    print("This question is related too device operations,,I am not supported to do")
```

```python
classify_query("set an alarm at 4'O clock")
```

step 3:: Check whether given question can be searched  or not in browser.

```python
def classify_query(query):
  get_from_web_search=False      #taking a boolean value.
  device_lst=['set an alarm','remainder','message','call']   #making a list of device operations
  device=False   #by default the device is set to be false.
  for i in device_lst:
    if i in query:
      device=True   #device will set to true.

  if device:   #condition....device==true
    print("this question is related too device things,,i am not supported to do")

  if device:
     get_from_web_search=False
  else :
     get_from_web_search=True
  return  get_from_web_search
```

```python
classify_query("set an alarm at 5pm")
```

this returns false because  it can be searched in internet .

```python
classify_query("what is the capital of India?")
```

This returns true coz this query can be searched in browser.

```python
import pathlib       #pathlib module provides easier way to manipulate file paths.
import textwrap       #textwrap module is used to format text for better readability.

import google.generativeai as genai      #importing  google.generativeai package alias genai

from IPython.display import display    #Ipython is a inbuilt jupyter library that imports display and markdown functions
from IPython.display import Markdown
#display function is used to show it properly
# markdown function is used for proper formating
def to_markdown(text):
  text = text.replace('•', '  *')     #replaces . with *,,, coz * is  a markdown syntax for creating bulleted lists by default
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

**step 5:: covert text to speech.**

```python
!pip  install gTTS    #installing gTTS package
```

```python
from gtts import gTTS     # importing gTTS module from gtts package
from IPython.display import Audio,display  #importing audio and display functions from module IPython.display
```

```python
def ask_gemini(query):
  modified_prompt="hey give me the answer to the quetion "+ query +" in 50 words "
  response=model.generate_content(modified_prompt)
  to_markdown(response.text)
  return response.text
```

```python
any_queries='y'
name=''
while any_queries=='y':
  if name=='':
      name=input("hey what is your name : ")
  q=ask_question(name)

  go_ahead=classify_query(q)
  answer=''
  if go_ahead==True:
      answer=ask_gemini(q)
  def speak(answer):

    tts=gTTS(answer)
    tts.save('answer.mp3')

    display(Audio('answer.mp3', autoplay=True))

  speak(answer)
  print(answer)


  any_queries=input("do you have any other questions? (type y/n)")
```
